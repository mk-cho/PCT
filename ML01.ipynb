```python
import pandas as pd

df = pd.read_excel('data.xlsx')
#print(df)
```

```python
df.head()
```

```python
df['d19']=df['d19'].astype('float')
df.info()
```

```python
Fracture = df.copy("R_critical")
Fracture = Fracture['R_critical'].apply(lambda x: 0 if x == 0.0 else 1).to_frame()
Fracture_new = Fracture.rename(columns={'R_critical':'Fractured'})
df_old = df
df = pd.concat([df,Fracture_new], axis=1)
#Fracture.info()
```

```python
df.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
import numpy as np

def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(df, 0.2)
print(len(train_set), "train +", len(test_set), "test")
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
print(len(train_set), "train +", len(test_set), "test")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
train_set.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
corr_matrix = df.corr(method='spearman')
corr_matrix['PCT'].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ['PCT', 'R_critical', "d1", 'd2', 'd7', 'd10','d16']
scatter_matrix(df[attributes], figsize=(12,8))
```

```python
df_train = train_set.drop(['PCT','R_critical','Fractured'],axis=1)
df_trainlabel = train_set['PCT'].copy()

df_test = test_set.drop(['PCT','R_critical','Fractured'],axis=1)
df_testlabel = test_set['PCT'].copy()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(df_train, df_trainlabel)
```

```python
print("예측", lin_reg.predict(df_test[:5]))
print("레이블", list(df_testlabel[:5]))
```

```python
from sklearn.metrics import mean_squared_error
predictions = lin_reg.predict(df_train)
lin_rmse = np.sqrt(mean_squared_error(df_trainlabel, predictions))
lin_rmse
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(df_train, df_trainlabel)
tree_predictions = tree_reg.predict(df_train)
tree_rmse = np.sqrt(mean_squared_error(df_trainlabel, tree_predictions))
tree_rmse
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_reg, df_train, df_trainlabel, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
print("scores:", tree_rmse_scores)
print("Mean:", tree_rmse_scores.mean())
print("SD:", tree_rmse_scores.std())
```

```python
lin_scores = cross_val_score(lin_reg, df_train, df_trainlabel, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)

print("scores:", lin_rmse_scores)
print("Mean:", lin_rmse_scores.mean())
print("SD:", lin_rmse_scores.std())
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(df_train, df_trainlabel)
forest_predictions = forest_reg.predict(df_train)
forest_rmse = np.sqrt(mean_squared_error(df_trainlabel, forest_predictions))
print("forest_rmse:", forest_rmse)

forest_scores = cross_val_score(forest_reg, df_train, df_trainlabel, scoring="neg_mean_squared_error", cv=10)
forest_rmse_scores = np.sqrt(-forest_scores)

print("scores:", forest_rmse_scores)
print("Mean:", forest_rmse_scores.mean())
print("SD:", forest_rmse_scores.std())
```

```python
from sklearn.svm import SVR

SVM_reg = SVR(kernel='rbf',
              #degree=3,
              #epsilon=0.1,
              #gamma='scale',
              #coef0=1,
              C=1)
SVM_reg.fit(df_train, df_trainlabel)
SVM_predictions = SVM_reg.predict(df_train)
SVM_rmse = np.sqrt(mean_squared_error(df_trainlabel, SVM_predictions))
print("SVM_rmse:", SVM_rmse)

SVM_scores = cross_val_score(SVM_reg, df_train, df_trainlabel, scoring="neg_mean_squared_error", cv=10)
SVM_rmse_scores = np.sqrt(-SVM_scores)

print("scores:", SVM_rmse_scores)
print("Mean:", SVM_rmse_scores.mean())
print("SD:", SVM_rmse_scores.std())
```
